```python
import os
import numpy as np
import h5py
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# ===================== 配置 =====================
DATA_DIR = r'dataset_v4'
CLASSES = ['Hammer', 'Knife', 'Pistol', 'Plier', 'Rifle',
           'Satellite_model', 'Snips', 'Stiletto', 'Wrench']
BATCH_SIZE = 4
MODEL_PATH = 'models/VGG19_CBAM_MultiLabel.h5'
THRESHOLD = 0.5
MAX_WRONG_SHOW = 20
# ================================================

# ===================== 🔥 强制修复H5编码（必加）=====================
def fix_h5_encoding(filename):
    try:
        with h5py.File(filename, 'r+') as f:
            for attr_name in ['keras_version', 'backend', 'tensorflow_version']:
                if attr_name in f.attrs:
                    val = f.attrs[attr_name]
                    if isinstance(val, str):
                        f.attrs[attr_name] = val.encode('utf8')
    except:
        pass

fix_h5_encoding(MODEL_PATH)

# ===================== CBAM 层（完全稳定版）=====================
class CBAM(tf.keras.layers.Layer):
    def __init__(self, ratio=8, **kwargs):
        super(CBAM, self).__init__(**kwargs)
        self.ratio = ratio

    def build(self, input_shape):
        self.channel_avg_pool = tf.keras.layers.GlobalAveragePooling2D()
        self.channel_max_pool = tf.keras.layers.GlobalMaxPooling2D()
        channel_dim = input_shape[-1]
        self.shared_mlp = tf.keras.Sequential([
            tf.keras.layers.Dense(channel_dim // self.ratio, activation='relu'),
            tf.keras.layers.Dense(channel_dim)
        ])
        self.conv2d = tf.keras.layers.Conv2D(1, (7,7), padding='same', activation='sigmoid')

    def call(self, inputs):
        avg_pool = self.channel_avg_pool(inputs)
        max_pool = self.channel_max_pool(inputs)
        avg_pool = tf.keras.layers.Reshape((1,1,-1))(avg_pool)
        max_pool = tf.keras.layers.Reshape((1,1,-1))(max_pool)
        
        channel_attention = tf.nn.sigmoid(
            self.shared_mlp(avg_pool) + self.shared_mlp(max_pool)
        )
        channel_refined = inputs * channel_attention

        avg_pool = tf.reduce_mean(channel_refined, axis=-1, keepdims=True)
        max_pool = tf.reduce_max(channel_refined, axis=-1, keepdims=True)
        concat = tf.keras.layers.Concatenate()([avg_pool, max_pool])
        spatial_attention = self.conv2d(concat)
        return channel_refined * spatial_attention

    def get_config(self):
        config = super(CBAM, self).get_config()
        config.update({"ratio": self.ratio})
        return config

# ===================== 🔥 核心：手动构建模型 + 手动加载权重 =====================
from tensorflow.keras.applications import VGG19
from tensorflow.keras import layers, models

def build_vgg19_with_cbam(num_classes):
    base_model = VGG19(weights='imagenet', include_top=False, input_shape=(224,224,3))
    for layer in base_model.layers:
        layer.trainable = False

    cbam_outputs = []
    for block_num in [1,2,3,4,5]:
        conv_num = 2 if block_num < 3 else 4
        layer_name = f'block{block_num}_conv{conv_num}'
        x = base_model.get_layer(layer_name).output
        x = CBAM(ratio=8)(x)
        cbam_outputs.append(x)

    adjusted_features = []
    for feat in cbam_outputs:
        s = int(feat.shape[1])
        while s > 7:
            feat = layers.MaxPooling2D(2)(feat)
            s = s // 2
        adjusted_features.append(feat)

    x = layers.Concatenate()(adjusted_features)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(256, activation='relu')(x)
    x = layers.Dropout(0.5)(x)
    outputs = layers.Dense(num_classes, activation='sigmoid')(x)
    return models.Model(base_model.input, outputs)

# 构建 + 加载权重（完全避开decode报错）
model = build_vgg19_with_cbam(9)
with h5py.File(MODEL_PATH, 'r') as f:
    root = f['model_weights'] if 'model_weights' in f else f
    for layer in model.layers:
        if layer.name in root:
            g = root[layer.name]
            weights = [g[w][()] for w in g.attrs['weight_names']]
            layer.set_weights(weights)

print("✅ 模型加载成功！")

# ===================== 加载验证集 =====================
val_df = pd.read_csv(os.path.join(DATA_DIR, 'val_labels.csv'))

def add_full_path(df, split):
    df['clean_filename'] = df['filename'].astype(str).str.replace('\\', '/', regex=False)
    df['full_filename'] = f"{split}/" + df['clean_filename']
    return df

val_df = add_full_path(val_df, 'val')
val_df = val_df[val_df['full_filename'].apply(lambda x: os.path.exists(os.path.join(DATA_DIR, x)))]

datagen = ImageDataGenerator(rescale=1./255)
val_generator = datagen.flow_from_dataframe(
    dataframe=val_df, directory=DATA_DIR, x_col='full_filename', y_col=CLASSES,
    target_size=(224,224), batch_size=BATCH_SIZE, class_mode='raw', shuffle=False
)

# ===================== 预测 =====================
y_true = []
y_pred = []
image_paths = []

for i in range(len(val_generator)):
    x, y = val_generator[i]
    p = model.predict(x, verbose=0)
    y_true.append(y)
    y_pred.append(p)
    s = i * BATCH_SIZE
    e = s + len(x)
    image_paths.extend(val_df['full_filename'].iloc[s:e].tolist())

y_true = np.vstack(y_true)
y_pred = np.vstack(y_pred)

# ===================== 筛选错误 =====================
wrong_indices = []
for idx in range(len(y_true)):
    true = set(np.where(y_true[idx] == 1)[0])
    pred = set(np.where(y_pred[idx] >= THRESHOLD)[0])
    if true != pred:
        wrong_indices.append(idx)

print(f"\n【验证集】错误总数：{len(wrong_indices)}")
if not wrong_indices:
    print("🎉 无错误！")
    exit()

# ===================== 绘制错误图 =====================
show_idx = wrong_indices[:MAX_WRONG_SHOW]
n = len(show_idx)

plt.figure(figsize=(16, 4 * n))
for i, idx in enumerate(show_idx):
    img_path = os.path.join(DATA_DIR, image_paths[idx])
    img = plt.imread(img_path)

    true_lab = [CLASSES[j] for j in np.where(y_true[idx]==1)[0]]
    pred_lab = [CLASSES[j] for j in np.where(y_pred[idx]>=THRESHOLD)[0]]

    plt.subplot(n, 1, i+1)
    plt.imshow(img)
    plt.axis('off')
    plt.title(f'true: {true_lab}\npred: {pred_lab}', color='red', fontsize=12)

plt.tight_layout()
plt.savefig('val_only_wrong.png', dpi=200, bbox_inches='tight')
plt.show()

# ===================== 保存错误列表 =====================
wrong_list = []
for idx in wrong_indices:
    tl = [CLASSES[j] for j in np.where(y_true[idx]==1)[0]]
    pl = [CLASSES[j] for j in np.where(y_pred[idx]>=THRESHOLD)[0]]
    wrong_list.append({
        'path': image_paths[idx],
        'true': ','.join(tl),
        'pred': ','.join(pl)
    })

pd.DataFrame(wrong_list).to_csv('val_only_wrong.csv', index=False, encoding='utf-8-sig')
print("✅ 全部完成！错误图片 + 表格已保存")
```
