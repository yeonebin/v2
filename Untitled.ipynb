```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.applications import VGG19
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import average_precision_score

# ===================== 配置区 =====================
DATA_DIR = r'dataset_v4'    # 确保这个路径正确！
CLASSES = ['Hammer', 'Knife', 'Pistol', 'Plier', 'Rifle',
           'Satellite_model', 'Snips', 'Stiletto', 'Wrench']
NUM_CLASSES = len(CLASSES)
BATCH_SIZE = 4
EPOCHS = 100
INITIAL_LR = 1e-4
MODEL_SAVE_PATH = 'models/VGG19_CBAM_MultiLabel.h5'
# =================================================

# 1. 定义 CBAM 模块
class CBAM(tf.keras.layers.Layer):
    def __init__(self, ratio=8, **kwargs):
        super(CBAM, self).__init__(**kwargs)
        self.ratio = ratio
        self.channel_axis = -1

    def build(self, input_shape):
        self.channel_avg_pool = layers.GlobalAveragePooling2D()
        self.channel_max_pool = layers.GlobalMaxPooling2D()
        channel_dim = input_shape[self.channel_axis]
        self.shared_mlp = tf.keras.Sequential([
            layers.Dense(channel_dim // self.ratio, activation='relu'),
            layers.Dense(channel_dim)
        ])
        self.conv2d = layers.Conv2D(1, (7, 7), padding='same', activation='sigmoid')
        super(CBAM, self).build(input_shape)

    def call(self, inputs):
        # 通道注意力
        avg_pool = self.channel_avg_pool(inputs)
        max_pool = self.channel_max_pool(inputs)
        avg_pool = layers.Reshape((1, 1, avg_pool.shape[-1]))(avg_pool)
        max_pool = layers.Reshape((1, 1, max_pool.shape[-1]))(max_pool)
        channel_attention = tf.nn.sigmoid(self.shared_mlp(avg_pool) + self.shared_mlp(max_pool))
        channel_refined = inputs * channel_attention

        # 空间注意力
        avg_pool = tf.reduce_mean(channel_refined, axis=self.channel_axis, keepdims=True)
        max_pool = tf.reduce_max(channel_refined, axis=self.channel_axis, keepdims=True)
        concat = layers.Concatenate()([avg_pool, max_pool])
        spatial_attention = self.conv2d(concat)
        return channel_refined * spatial_attention

    def get_config(self):
        config = super(CBAM, self).get_config()
        config.update({"ratio": self.ratio})
        return config

# 2. 构建 VGG19 + CBAM 多标签模型
def build_vgg19_with_cbam(num_classes):
    base_model = VGG19(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
    for layer in base_model.layers:
        layer.trainable = False

    cbam_outputs = []
    for block_num in [1, 2, 3, 4, 5]:
        conv_num = 2 if block_num < 3 else 4
        layer_name = f'block{block_num}_conv{conv_num}'
        x = base_model.get_layer(layer_name).output
        x = CBAM(ratio=8)(x)
        cbam_outputs.append(x)

    target_size = 7
    adjusted_features = []
    for feature in cbam_outputs:
        current_size = int(feature.shape[1])
        downsample_factor = current_size // target_size
        x = feature
        if downsample_factor > 1:
            num_pools = int(np.log2(downsample_factor))
            for _ in range(num_pools):
                x = layers.MaxPooling2D(2)(x)
        adjusted_features.append(x)

    fused_feature = layers.Concatenate()(adjusted_features)
    x = layers.GlobalAveragePooling2D()(fused_feature)
    x = layers.Dense(256, activation='relu')(x)
    x = layers.Dropout(0.5)(x)
    outputs = layers.Dense(num_classes, activation='sigmoid')(x)
    model = models.Model(inputs=base_model.input, outputs=outputs)
    return model

# 3. 自定义回调：每个 epoch 后计算验证集 mAP
class MultiLabelMAP(tf.keras.callbacks.Callback):
    def __init__(self, validation_data):
        super().__init__()
        self.validation_data = validation_data

    def on_epoch_end(self, epoch, logs=None):
        steps = len(self.validation_data)
        all_y_true = []
        all_y_pred = []
        for _ in range(steps):
            x_val, y_val = next(self.validation_data)
            preds = self.model.predict(x_val, verbose=0)
            all_y_true.append(y_val)
            all_y_pred.append(preds)
        y_true = np.vstack(all_y_true)
        y_pred = np.vstack(all_y_pred)

        aps = []
        for i in range(NUM_CLASSES):
            if np.sum(y_true[:, i]) > 0:
                ap = average_precision_score(y_true[:, i], y_pred[:, i])
                aps.append(ap)
        mAP = np.mean(aps) if aps else 0.0
        print(f" — val_mAP: {mAP:.4f}")
        logs['val_mAP'] = mAP

# 4. 加载数据并构建完整路径 + 【新增：文件存在性检查+调试】
train_df = pd.read_csv(os.path.join(DATA_DIR, 'train_labels.csv'))
val_df = pd.read_csv(os.path.join(DATA_DIR, 'val_labels.csv'))

# 🔥 终极修复版：彻底解决 \ 和 / 路径分隔符问题
def add_full_path(df, split):
    """
    核心修复：
    1. 统一把所有 \ 替换为 /（Linux 标准路径）
    2. 直接使用 CSV 里的完整文件名（已包含子目录）
    """
    # 第一步：把 CSV 里的反斜杠 \ 全部替换成正斜杠 /
    df['clean_filename'] = df['filename'].astype(str).str.replace('\\', '/', regex=False)
    
    # 第二步：拼接最终路径（split + 清理后的文件名）
    df['full_filename'] = f"{split}/" + df['clean_filename']
    
    # 检查文件是否存在
    missing_files = []
    for idx, path in enumerate(df['full_filename']):
        full_path = os.path.join(DATA_DIR, path)
        if not os.path.exists(full_path):
            missing_files.append(full_path)
    
    if len(missing_files) > 0:
        print(f"\n⚠️  【{split}集】缺失 {len(missing_files)} 张图片：")
        for f in missing_files[:5]:
            print(f"缺失：{f}")
        print(f"✅ 正确路径格式：{split}/Multiple_Targets/xxx.jpg\n")
    else:
        print(f"\n✅ 【{split}集】所有图片路径正常！")
    
    # 打印前3个路径验证
    print(f"📌 生成的标准路径示例：{df['full_filename'].head(3).to_list()[0]}")
    
    return df

# 处理路径
train_df = add_full_path(train_df, 'train')
val_df = add_full_path(val_df, 'val')

# 数据生成器（仅归一化）
datagen = ImageDataGenerator(rescale=1./255)

# 【关键：过滤掉不存在的文件】
train_df = train_df[train_df['full_filename'].apply(lambda x: os.path.exists(os.path.join(DATA_DIR, x)))]
val_df = val_df[val_df['full_filename'].apply(lambda x: os.path.exists(os.path.join(DATA_DIR, x)))]

print(f"\n📊 过滤后有效样本：")
print(f"训练集：{len(train_df)} 张")
print(f"验证集：{len(val_df)} 张")

# 如果没有有效样本，直接退出
if len(train_df) == 0 or len(val_df) == 0:
    raise Exception("❌ 没有找到任何有效图片！请检查数据集路径/结构/文件名")

# 构建生成器
train_generator = datagen.flow_from_dataframe(
    dataframe=train_df,
    directory=DATA_DIR,
    x_col='full_filename',
    y_col=CLASSES,
    target_size=(224, 224),
    batch_size=BATCH_SIZE,
    class_mode='raw',
    shuffle=True
)

val_generator = datagen.flow_from_dataframe(
    dataframe=val_df,
    directory=DATA_DIR,
    x_col='full_filename',
    y_col=CLASSES,
    target_size=(224, 224),
    batch_size=BATCH_SIZE,
    class_mode='raw',
    shuffle=False
)

print(f"\n✅ 生成器批次：")
print(f"训练集批次：{len(train_generator)}")
print(f"验证集批次：{len(val_generator)}")

# 5. 构建模型并编译
os.makedirs('models', exist_ok=True)  # 自动创建模型保存文件夹
with tf.device('/GPU:0' if tf.config.list_physical_devices('GPU') else '/CPU:0'):
    model = build_vgg19_with_cbam(NUM_CLASSES)
    model.compile(
        optimizer=Adam(learning_rate=INITIAL_LR),
        loss='binary_crossentropy',
        metrics=['binary_accuracy']
    )

    # 回调函数
    map_callback = MultiLabelMAP(validation_data=val_generator)
    checkpoint = tf.keras.callbacks.ModelCheckpoint(
        MODEL_SAVE_PATH, monitor='val_loss', mode='min', save_best_only=True, verbose=1
    )
    early_stop = tf.keras.callbacks.EarlyStopping(
        monitor='val_loss', patience=20, mode='min', restore_best_weights=True, verbose=1
    )
    reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(
        monitor='val_loss', factor=0.2, patience=5, min_lr=1e-6, verbose=1
    )

    # 训练
    history = model.fit(
        train_generator,
        steps_per_epoch=len(train_generator),
        epochs=EPOCHS,
        validation_data=val_generator,
        validation_steps=len(val_generator),
        callbacks=[map_callback, checkpoint, early_stop, reduce_lr],
        verbose=1
    )

# 6. 保存模型
model.save(MODEL_SAVE_PATH)
print(f"\n✅ 模型已保存至：{MODEL_SAVE_PATH}")

# 7. 可视化训练曲线
plt.figure(figsize=(20, 5))

plt.subplot(1, 3, 1)
plt.plot(history.history['binary_accuracy'], label='Train Accuracy')
plt.plot(history.history['val_binary_accuracy'], label='Val Accuracy')
plt.title('Binary Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

if 'val_mAP' in history.history:
    plt.subplot(1, 3, 3)
    plt.plot(history.history['val_mAP'], label='Val mAP', color='green')
    plt.title('Mean Average Precision')
    plt.xlabel('Epoch')
    plt.ylabel('mAP')
    plt.legend()

plt.tight_layout()
plt.savefig('training_curves.png', dpi=300)
plt.show()

print("\n✅ 训练完成！曲线已保存为 training_curves.png")
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from PIL import Image

# ===================== 【配置】只用你的现成模型 =====================
DATA_DIR = r'dataset_v4'
CLASSES = ['Hammer', 'Knife', 'Pistol', 'Plier', 'Rifle', 'Satellite_model', 'Snips', 'Stiletto', 'Wrench']
MODEL_PATH = 'models/VGG19_CBAM_MultiLabel.h5'
THRESHOLD = 0.5
BATCH_SIZE = 4

# ===================== 🔥 终极修复：直接屏蔽TF内部decode报错 =====================
import tensorflow.python.keras.saving.hdf5_format as hdf
old_load = hdf.load_model_from_hdf5

def fixed_load(filepath, custom_objects=None, compile=True):
    import h5py
    f = h5py.File(filepath, 'r')
    
    # 直接绕过报错的那一行！！！
    model_config = f.attrs.get('model_config')
    if isinstance(model_config, bytes):
        model_config = model_config.decode('utf-8')
    
    from tensorflow.python.keras.saving import model_config
    model = model_config.model_from_json(model_config, custom_objects=custom_objects)
    
    # 跳过版本检查，直接加载权重
    hdf.load_weights_from_hdf5_group(f['model_weights'], model.layers)
    if compile:
        from tensorflow.python.keras.saving import compile_utils
        compile_utils.compile_from_config(model, f)
    return model

# 强制替换TF报错函数
hdf.load_model_from_hdf5 = fixed_load

# ===================== CBAM必须写（你模型里有这个层）=====================
class CBAM(tf.keras.layers.Layer):
    def __init__(self, ratio=8, **kwargs):
        super(CBAM, self).__init__(**kwargs)
        self.ratio = ratio
    def build(self, input_shape):
        self.channel_avg_pool = tf.keras.layers.GlobalAveragePooling2D()
        self.channel_max_pool = tf.keras.layers.GlobalMaxPooling2D()
        channel_dim = input_shape[-1]
        self.shared_mlp = tf.keras.Sequential([
            tf.keras.layers.Dense(channel_dim//self.ratio, activation='relu'),
            tf.keras.layers.Dense(channel_dim)
        ])
        self.conv2d = tf.keras.layers.Conv2D(1, (7,7), padding='same', activation='sigmoid')
    def call(self, inputs):
        avg = tf.keras.layers.Reshape((1,1,-1))(self.channel_avg_pool(inputs))
        max = tf.keras.layers.Reshape((1,1,-1))(self.channel_max_pool(inputs))
        ca = tf.nn.sigmoid(self.shared_mlp(avg) + self.shared_mlp(max))
        x = inputs * ca
        sa = self.conv2d(tf.keras.layers.Concatenate()([tf.reduce_mean(x,-1,keepdims=True), tf.reduce_max(x,-1,keepdims=True)]))
        return x * sa
    def get_config(self):
        config = super(CBAM, self).get_config()
        config.update({"ratio": self.ratio})
        return config

# ===================== ✅ 直接加载你的现成模型 =====================
print("加载模型中...")
model = tf.keras.models.load_model(MODEL_PATH, custom_objects={"CBAM": CBAM})
print("✅ 模型加载成功！")

# ===================== 加载验证集 =====================
val_df = pd.read_csv(os.path.join(DATA_DIR, 'val_labels.csv'))

def add_path(df, split):
    df['full'] = f"{split}/" + df['filename'].astype(str).str.replace('\\','/')
    return df

val_df = add_path(val_df, 'val')
val_df = val_df[val_df['full'].apply(lambda x: os.path.exists(os.path.join(DATA_DIR, x)))]

# 生成器
gen = ImageDataGenerator(rescale=1./255)
val_gen = gen.flow_from_dataframe(
    val_df, DATA_DIR, x_col='full', y_col=CLASSES,
    target_size=(224,224), batch_size=BATCH_SIZE,
    class_mode='raw', shuffle=False
)

# ===================== 预测 =====================
y_true = []
y_pred = []
paths = []

for i in range(len(val_gen)):
    x, y = val_gen[i]
    p = model.predict(x, verbose=0)
    y_true.append(y)
    y_pred.append(p)
    paths.extend(val_df['full'].iloc[i*BATCH_SIZE : i*BATCH_SIZE+len(x)])

y_true = np.vstack(y_true)
y_pred = np.vstack(y_pred)

# ===================== 找错误 =====================
wrong = [i for i in range(len(y_true))
         if set(np.where(y_true[i]==1)[0]) != set(np.where(y_pred[i]>=THRESHOLD)[0])]

print(f"\n错误数量：{len(wrong)}")

# ===================== 画图 =====================
if wrong:
    plt.figure(figsize=(16, 4*len(wrong[:20])))
    for i, idx in enumerate(wrong[:20]):
        img = Image.open(os.path.join(DATA_DIR, paths[idx]))
        t = [CLASSES[j] for j in np.where(y_true[idx]==1)[0]]
        p = [CLASSES[j] for j in np.where(y_pred[idx]>=THRESHOLD)[0]]
        plt.subplot(len(wrong[:20]), 1, i+1)
        plt.imshow(img)
        plt.axis('off')
        plt.title(f'TRUE: {t} | PRED: {p}', color='r')
    plt.tight_layout()
    plt.savefig('val_wrong.png', dpi=200)
    plt.show()
    print("✅ 错误图片已保存！")
else:
    print("🎉 无错误！")
```
