```python
# unzip_file.py
import zipfile
import os

def main():
    # 配置参数（可根据需要修改）
    ZIP_FILE_PATH = "dataset_v4.zip"  # 压缩包名称
    EXTRACT_TO = "./"                     # 解压目标路径（当前目录）

    # 检查压缩包是否存在
    if not os.path.exists(ZIP_FILE_PATH):
        print(f"❌ 错误：压缩包 {ZIP_FILE_PATH} 不存在！")
        return

    # 创建解压目录
    os.makedirs(EXTRACT_TO, exist_ok=True)

    # 解压核心逻辑
    try:
        with zipfile.ZipFile(ZIP_FILE_PATH, 'r') as zip_ref:
            # 打印压缩包内容
            file_list = zip_ref.namelist()
            print(f"📦 压缩包 {ZIP_FILE_PATH} 内容：")
            for idx, file in enumerate(file_list):
                if idx < 5:  # 只显示前5个文件
                    print(f"  {idx+1}. {file}")
                elif idx == 5:
                    print(f"  ... 还有 {len(file_list)-5} 个文件未显示")
            
            # 执行解压
            zip_ref.extractall(EXTRACT_TO)
            print(f"\n✅ 解压成功！文件已解压到：{os.path.abspath(EXTRACT_TO)}")
    
    except zipfile.BadZipFile:
        print(f"❌ 错误：{ZIP_FILE_PATH} 不是有效的ZIP文件（可能损坏/格式错误）")
    except PermissionError:
        print(f"❌ 错误：无权限写入 {EXTRACT_TO} 目录，请检查权限")
    except Exception as e:
        print(f"❌ 解压失败：{str(e)}")

if __name__ == "__main__":
    main()
```
